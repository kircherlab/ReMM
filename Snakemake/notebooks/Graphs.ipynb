```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

x = np.array([ 1.92, 30, 34.21])
y = np.array([7.30, 5, 0.06])

def fun(x, a, b, c):
    return a * np.cosh(b * x )+ c


def pl(x,y,name):
    coef,_ = curve_fit(fun, x, y)
    
    a = np.linspace(x[0],x[-1])
    b = fun(np.linspace(x[0],x[-1]), *coef)
    plt.plot(a/a.max(),b/b.max(), label=name,linewidth=3)


s = 3
fig, ax = plt.subplots(figsize=(16,9))  
    
pl(x,y,'PRC 1')

y = np.array([7.30, 4, 0.06])

pl(x,y,'PRC 2')

y = np.array([7.30, 3, 0.06])
pl(x,y,'PRC 3')


#y = np.array([7.30, 1.8, 0.06])
#pl(x,y,'PRC 4')
#ax.plot([0,1],[1,1],[1,1],[1,0],color = 'red')
ax.plot([0.055,1],[0.01,0.01],color = 'red', linestyle='dashed',label='Random classifier')



ax.tick_params(axis='both' , labelsize=8*s)

ax.set_xlabel('Recall', fontsize = 8*s,labelpad=10)
ax.set_ylabel('Presicion',fontsize = 8*s,labelpad=10)

ax.plot(1,1,'ob', linewidth=10, markersize=18,marker='*',color = 'black',label='Perfect classifier')
#ax.set_xticks(np.linspace(0, 1, 7).round(1))
plt.legend(fontsize = 8*s,prop={'size': 25})
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(1)
plt.savefig('figures/prc_example.png',di=30)

plt.show()
```

```python

def pl1(xData,yData,name):
    import numpy, scipy, matplotlib
    import matplotlib.pyplot as plt
    from scipy.optimize import curve_fit




    def func(x, a, b, offset): # polytrope equation from zunzun.com
        return  a / numpy.power(x, b) + offset


    # these are the same as the scipy defaults
    initialParameters = numpy.array([1.0, 1.0, 1.0])

    # curve fit the test data
    fittedParameters, pcov = curve_fit(func, xData, yData, initialParameters)

    modelPredictions = func(xData, *fittedParameters) 

    absError = modelPredictions - yData
    xModel = numpy.linspace(min(xData), max(xData), 1000)
    yModel = func(xModel, *fittedParameters)
    plt.plot(xModel/xModel.max(),yModel/yModel.max(),label=name,linewidth = 3)

    return xModel/xModel.max(),yModel/yModel.max()
   
```

```python





s = 3
fig, ax = plt.subplots(figsize=(16,9))  
    


#y = np.array([7.30, 1.8, 0.06])
#pl(x,y,'PRC 4')
#ax.plot([0,1],[1,1],[1,1],[1,0],color = 'red')



xData = np.array([7e-09, 9e-09, 1e-08, 2e-8, 1e-6])
yData = np.array([790.0, 870.0, 2000.0, 2000.0, 3100.0])
pl1(xData,yData,'AUC 1')

xData = np.array([7e-09, 9e-09, 1e-08, 2e-8, 1e-6])
yData = np.array([790.0, 600.0, 100.0, 1500.0, 3100.0])
pl1(xData,yData,'AUC 3')

xData = np.array([7e-09, 9e-09, 1e-08, 2e-8, 1e-6])
yData = np.array([790.0, 700.0, 800.0, 2000.0, 3100.0])
pl1(xData,yData,'AUC 2')
#ax.plot([0.055,1],[1,0.01],color = 'red', linestyle='dashed',label='Chance')



ax.tick_params(axis='both' , labelsize=8*s)

ax.set_xlabel('False Positive Rate', fontsize = 8*s,labelpad=10)
ax.set_ylabel('True Positive Rate',fontsize = 8*s,labelpad=10)

ax.plot(0,1, linewidth=10, markersize=10,marker='o',color = 'black',label='Perfect classifier')
#ax.set_xticks(np.linspace(0, 1, 7).round(1))
plt.legend(fontsize = 8*s,prop={'size': 25})
#plt.xscale('log')

plt.show()
```
